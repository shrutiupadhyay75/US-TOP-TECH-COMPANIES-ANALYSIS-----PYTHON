**Importing Dependencies**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Creating "Tech" DataFrame for tech companies**

```python
tech=pd.read_csv("D:\Self-Projects\Top Companies\Top 50 US Tech Companies 2022 - 2023.csv")
```

**Checking information about the Dataset**

```python
tech.info()
```

```python
#Checking for null values
tech.isnull().sum()
```

```python
tech.describe()
```

**Checking for rows and columns using shape attribute**

```python
tech.shape
```

**Total company sectors**

```python
tech['Sector'].value_counts()
```

```python
plt.figure(figsize=(20,10))
plt.title("The most technology company sectors")
sns.countplot(x='Sector', data=tech, order=tech['Sector'].value_counts().index)
plt.xlabel("Sector")
plt.ylabel("Count of Sector")
plt.show()
```

**Headquarter states**

```python
tech['HQ State'].value_counts()
```

```python
plt.figure(figsize=(18,8))
plt.title("The most technology company in states")
sns.countplot(x='HQ State', data=tech, order=tech['HQ State'].value_counts().index)
plt.xlabel("State")
plt.ylabel("Count of State")
```

**TOP 10 companies with higher revenue**

```python
top_10 = tech[['Company Name', 'Annual Revenue 2022-2023 (USD in Billions)']].nlargest(10,'Annual Revenue 2022-2023 (USD in Billions)')
top_10
```

```python
plt.figure(figsize=(20,8))
plt.title("Top 10 Highest Tech Company Annual Revenue Between 2022 to 2023")
sns.barplot(x=top_10['Company Name'], y=top_10['Annual Revenue 2022-2023 (USD in Billions)'])
plt.xlabel("Company")
plt.ylabel("Annual Revenue (USD in Billions)")
```

**Top 10 highest Market Caps**

```python
top_10_cap= tech[['Company Name', 'Market Cap (USD in Trillions)']].nlargest(10,'Market Cap (USD in Trillions)')
top_10_cap
```

```python
plt.figure(figsize=(24,8))
plt.title("Top 10 Highest Tech Company Market Cap Between 2022 to 2023")
sns.barplot(x=top_10_cap['Company Name'], y=top_10_cap['Market Cap (USD in Trillions)'])
plt.xlabel("Company")
plt.ylabel("Market Cap (USD in Trillions)")
```

**Top 10 highest annual income tax**

```python
top_10_tax = tech[['Company Name', 'Annual Income Tax in 2022-2023 (USD in Billions)']]\
.nlargest(10,'Annual Income Tax in 2022-2023 (USD in Billions)')
top_10_tax
```

```python
plt.figure(figsize=(24,10))
plt.title("Top 10 Highest Tech Company Annual Income Tax Between 2022 to 2023")
sns.barplot(x=top_10_tax['Company Name'], y=top_10_tax['Annual Income Tax in 2022-2023 (USD in Billions)'])
plt.xlabel("Company")
plt.ylabel("Annual Income Tax (USD in Billions)")
```

**Oldest technology company**

```python
tech[tech['Founding Year'] == tech['Founding Year'].min()]
```

**Newest technology company**

```python
tech[tech['Founding Year'] == tech['Founding Year'].max()]
```

**Top 10 highest number of employees in each company**

```python
top_emp = tech[['Company Name', 'Employee Size']].nlargest(10,'Employee Size')
top_emp
```

```python
plt.figure(figsize=(20,8))
plt.title("Top 10 Highest Employee Amount")
sns.barplot(x=top_emp['Company Name'], y=top_emp['Employee Size'])
plt.xlabel("Company")
plt.ylabel("Number of Employee")
```
